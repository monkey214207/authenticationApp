```python
import pandas
import matplotlib.pyplot as plt
```

```python
dir = '10669_session_8/'
TouchEvent_colnames = ['Systime','EventTime','ActivityID','Pointer_count','PionterID','ActionID','X','Y','Pressure','Contact_size','Phone_orientation']
TouchEvent = pandas.read_csv(dir+'TouchEvent.csv', names=TouchEvent_colnames)
Systime_touchevent = TouchEvent.Systime.tolist()
ActionID = TouchEvent.ActionID.tolist()
TouchEvent
```

```python
t_start = Systime_touchevent[0] 
t_end = Systime_touchevent[-1]   
t_100msBefore = t_start - 100 
t_100msAfter = t_end + 100
t_200msAfter = t_end + 200
```

```python
Accelerometer_colnames = ['Systime','EventTime','ActivityID','X','Y','Z','Phone_orientation']
Accelerometer = pandas.read_csv(dir+'Accelerometer.csv', names=Accelerometer_colnames)
Z = Accelerometer.Z.tolist()
Systime_Acc = Accelerometer.Systime.tolist()
Accelerometer
```

```python
plt.plot(Systime_Acc,Z)
```

```python
before100ms = []
after200ms = []
for i in Systime_Acc:
    if i>t_100msBefore and i<t_start:
        before100ms.append(Systime_Acc.index(i))
    if i>t_end and i<t_200msAfter:
        after200ms.append(Systime_Acc.index(i))
print(before100ms)
print(after200ms, len(after200ms))
```

```python
avg100msBefore = sum(Z[before100ms[0]:before100ms[-1]+1]) / len(before100ms)
avg100msBefore 
```

```python
diff = []
for i in after200ms:
    diff.append(abs(Z[i]-avg100msBefore))
print(diff)
print(len(diff))
```

```python
avgDiff = []
for i in range(len(diff)):
    avgDiff.append(sum(diff[0:i+1])/(i+1))
print(avgDiff, len(avgDiff))
```

```python
print(min(avgDiff))
print(avgDiff.index(min(avgDiff)))
```
